# Baseline Forecasts

Forecasts are scaled to the revenue and growth rate projections in the FY21 to FY25 Five Year Plan, as proposed on March 5, 2020.

Forecasts are calculated for: 

- [Wage Tax](#wage)
- [Sales Tax](#sales)
- [BIRT](#birt)
- [Realty Transfer Tax](#rtt)
- [Parking Tax](#parking)
- [Amusement Tax](#amusement)
- [Net Profits Tax](#sales)
- [Soda Tax](#soda)

```python
from covid19_forecaster.baseline import RevenueForecast
```

```python
OUTPUT_COLUMNS = ['date', 'year', 'month', 'fiscal_year', 'total', 'naics_sector']
```

## Wage Tax <a id="wage"></a>

```python
wage = RevenueForecast("wage")
```

```python
fig, ax = wage.plot(kind='calibrated')
```

```python
# Save
wage.forecast[OUTPUT_COLUMNS].to_csv("../projections/wage.csv", index=False)
```

```python
wage.forecast.groupby("fiscal_year")['total'].sum() / 1e6
```
